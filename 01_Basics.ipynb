### Tuples and their basic manipulation

```python
#Tuples as an object type that does not allow modification of its content inmutable
tuple1 = (0, 1, 2, 3, 4, 3, 2, 1)
tuple2 = (10, True, "A", 5.0)

print((type(tuple1), tuple1), #Object type and content
	  len(tuple1), #Length
	  sum(tuple1), #Summation of its elements
	  tuple1.count(3), #Counting the appearances of a particular element
	  tuple1.index(4), #Finding the position of an element
	  tuple1[0:5], #Indexing
	  tuple1 + tuple2, #Combining two tuples
	  sep="\n")
#Tuples do not allow for vectorized operations nor scalar transformations. Not so ever, it is important to highlight that they allow for different data types.

#Deleting the temporary variables from the memory
del tuple1, tuple2
```

### Lists and their basic manipulation

```python
#List object as a versatile structure that allows manipulating its content
list1 = [0, 1, 2, 3, 4, 3, 2, 1]
list2 = [10, True, "A", 5.0]
list3 = list(np.random.normal(0, 100, 10))

print((type(list1), list1), #Object type and content
	  len(list1), #Length
	  sum(list1), #Summation of its elements
	  list1.count(3), #Counting the appearances of a particular element
	  list1.index(4), #Finding the position of an element
	  list1[0:5], #Indexing,
	  list1 * 2, #Duplicating elements
	  list1 + list2, #Combining two tuples
	  list1.clear(), #Cleaning all content (Modifies the original list if a copy is not created)
	  sep="\n")
#Lists do not allow for vectorized operations nor scalar transformations. Not so ever, it is important to highlight that they allow for different data types.

#Performing additinal manipulations that would alter the original list in case that a copy wass not created.
list1.append("Hello") #Adding a single element to the empty list.
print(list1)
list1.extend([i for i in range(0, 10, 2)]) #Adding multiple elements to a list
print(list1)
list1[3] = "Replacement" #Substituting an element
print(list1)
list1.insert(3, ["Inserted", 8]) #Inserting an element without substitution in this case a sublist
print(list1)
list1.pop(0) #Deleting an item
list1.extend([2]* 5) #Inserting a n times object
list1.remove(2) #Removing the first occurance
print(list1)
list1.reverse() #Inverting the order
print(list1)
list3.sort() #Sorting in ascending order elements of a list
print(list3)

#Deleting the temporary variables from the memory
del list1, list2, list3
```

### Sets and their basic manipulation

```python
#Sets as elements that allows unique elements, analogous to its mathematical principle
set1 = {1, 2, 3, 1, 2, 3, 1, 2, 3, 4, 5}
set2 = {3, 4, 5, 6, 5, 6, 7, 8}

print((type(set1), set1), #Object type and content
	  len(set1), #Length
	  sum(set1), #Summation of its elements
	  set1.difference(set2), #Getting the elements that are different in the first set compared to the second (antileft join)
	  set1.intersection(set2), #Common elements between sets (join)
	  set1.isdisjoint(set2), #Capturing if the two elements have a null intersection
	  set1.symmetric_difference(set2), #Getting all elements that are distincts in the two sets (anti outer join)
	  set1.union(set2), #Unification of the two sets with the unique components
	  sep="\n")
#Sets do not allow for vectorized operations nor scalar transformations. Not so ever, it is important to highlight that they allow for different data types.

set1.add("Hello") #Adding a new element
print(set1)
set1.pop() #Dropping out the first element
print(set1)
set1.remove("Hello") #Dropping an specifical element form the set
print(set1)

#Deleting the temporary variables from the memory
del set1, set2
```

## Data simulation

```python
#Creating an array for basic plotting
np.random.seed(300) #Setting the seed for reproducibility
simulated_data = np.column_stack([np.random.normal(0, 1, 100), np.random.binomial(10, 0.5, 100)]) #Simulating the data with two variables, one discrete and another continuous

#Preview
print(simulated_data[:5,], simulated_data[-5:,], sep="\n\n")
```

```python
#Measurements of central tendency manual calculus
for column in range(0, simulated_data.shape[1]):
	print("Mean: {}, median: {}, variance: {}, skewness: {}, kurtosis: {}".format(	
		np.mean(simulated_data[:,column]),
		np.median(simulated_data[:,column]),
		np.var(simulated_data[:,column]),
		scipy.stats.skew(simulated_data[:,column], axis=0, bias=False),
		scipy.stats.kurtosis(simulated_data[:,column], axis=0, fisher=False, bias=False)))
	
print(f"\nCorrelation coefficient: \n{np.corrcoef(simulated_data[:, 0], simulated_data[:, 1])}")
```

According to our random simulated process, and confirmed by the descriptive statistics, the two variables are not linearly correlated.

```python
#Exploratory plot
plt.plot(simulated_data[:, 0], simulated_data[:, 1])
plt.title("Exploratory plot variable 1 vs variable 2")
plt.show()

#Basic plots for understanding their distribution and concentration
fig, axs = plt.subplots(2, 2)

axs[0, 0].set_title('Strip plot for first column')
axs[0, 0].plot(simulated_data[:,0], "ro")
axs[0, 1].set_title('Histogram for first column')
axs[0, 1].hist(simulated_data[:,0], facecolor='r')

axs[1, 0].set_title('Strip plot for second column')
axs[1, 0].plot(simulated_data[:,1], "bo")
axs[1, 1].set_title('Histogram for second column')
axs[1, 1].hist(simulated_data[:,1])

fig.tight_layout()
```

With the different plots we can see that our sample has a randomized behaviour in which there is not a discernible pattern between our two variables. And we can note that both tend to have the normal and binomial expected behaviour.
