### Loops and control structures

```python
##Demonstration via a figure
#Creating a square with an X figure in the center by loops and control structures.
#This feature only works efficiently in squares, differences in the max and row columns could cause asymmetries in the inner figure.
#For even numbers of rows and columns it retrieves a square with diagonals without a connection point. For odd numbers, it retrieves two diagonals connected by a centroid.
max_row = 7
max_col = 7
for i in range(0, max_row): #Iterator for rows
	for j in range(0, max_col): #Iterator for columns
		if ((j==0) | (j==max_col-1)) | ((i==0) | (i==max_row-1)) | ((i==j) | (i+j==max(max_row, max_col)-1)):
			print("*", end="\t")
		else:
			print("", end="\t")
	print("\n")

#Deleting the temporary variables from the memory
del max_row, max_col
```

```python
## Demonstration via the calculus of the pascal triangle
maximum_level = 6

for i in range(1, maximum_level+1): #Iterator for rows
    constant = 1 #Right diagonal element
    for j in range(0, maximum_level-i+1): #Iterator for columns, displacement of the elements
        print('\t', end='')

    for j in range(1, i+1): #Iterator for columns, emphasizing in filling the values
        #First  element of the sequence is 1
        print('\t', constant, sep='\t', end='')
        # Calculating the binomial coefficient for inner values of the triangle
        constant = (constant * (i - j)) // j
    print("\n")

#Deleting the temporary variables from the memory
del constant, maximum_level
```

```python
#Applying a exponential random decreasing to the squared value until reaching the 0
value = 1000000 #Stablishsing a starting point
while value >= 0:
	try:
		value = np.sqrt(value) - np.random.exponential(scale=5, size=1)[0] #Obtaining a random number from the exponential distribution
		if value >= 0:
			print(value)
	except:
		break

#Deleting the temporary variables from the memory
del value
```

```python
#Calculating a linear regression via OLS by the algebraic representation
yvalues = np.transpose(np.array(np.random.normal(5, 5, 20), ndmin=2)) #Generating random numbers
xvalues = np.column_stack([np.random.binomial(30, 0.5, 20)]) #Generating random explanatory variables

#Restructuring X to add the constant term
xvalues_ols = np.column_stack([np.ones((xvalues.shape[0], 1)), xvalues])

#Calculating the coefficients via linear algebra
estimates = np.dot(np.linalg.inv(np.dot(np.transpose(xvalues_ols), xvalues_ols)), np.dot(np.transpose(xvalues_ols), yvalues))
yestim = np.dot(xvalues_ols, estimates)

print("Intercept: {} \nParameter B1: {}".format(estimates[0], estimates[1]))
#Plotting the random dots and the regression line
plt.title("Regression")
plt.scatter(xvalues, yvalues, label="Random dots")
plt.plot(xvalues, yestim, c="r", label="Regression line")
plt.legend(loc="upper left")
plt.show()

plt.title("Residuals")
plt.scatter(x=range(1, len(yvalues)+1), y=(yestim - yvalues))
plt.plot(range(1, len(yvalues)+1), np.array([0]* (len(yvalues))), "--")
plt.show()

#Deleting the temporary variables from the memory
del xvalues, xvalues_ols, yvalues, yestim, estimates
```

## Data simulation

```python
#Creating an array for basic plotting
np.random.seed(300) #Setting the seed for reproducibility
simulated_data = np.column_stack([np.random.normal(0, 1, 100), np.random.binomial(10, 0.5, 100)]) #Simulating the data with two variables, one discrete and another continuous

#Preview
print(simulated_data[:5,], simulated_data[-5:,], sep="\n\n")
```

```python
#Measurements of central tendency manual calculus
for column in range(0, simulated_data.shape[1]):
	print("Mean: {}, median: {}, variance: {}, skewness: {}, kurtosis: {}".format(	
		np.mean(simulated_data[:,column]),
		np.median(simulated_data[:,column]),
		np.var(simulated_data[:,column]),
		scipy.stats.skew(simulated_data[:,column], axis=0, bias=False),
		scipy.stats.kurtosis(simulated_data[:,column], axis=0, fisher=False, bias=False)))
	
print(f"\nCorrelation coefficient: \n{np.corrcoef(simulated_data[:, 0], simulated_data[:, 1])}")
```

According to our random simulated process, and confirmed by the descriptive statistics, the two variables are not linearly correlated.

```python
#Exploratory plot
plt.plot(simulated_data[:, 0], simulated_data[:, 1])
plt.title("Exploratory plot variable 1 vs variable 2")
plt.show()

#Basic plots for understanding their distribution and concentration
fig, axs = plt.subplots(2, 2)

axs[0, 0].set_title('Strip plot for first column')
axs[0, 0].plot(simulated_data[:,0], "ro")
axs[0, 1].set_title('Histogram for first column')
axs[0, 1].hist(simulated_data[:,0], facecolor='r')

axs[1, 0].set_title('Strip plot for second column')
axs[1, 0].plot(simulated_data[:,1], "bo")
axs[1, 1].set_title('Histogram for second column')
axs[1, 1].hist(simulated_data[:,1])

fig.tight_layout()
```

With the different plots we can see that our sample has a randomized behaviour in which there is not a discernible pattern between our two variables. And we can note that both tend to have the normal and binomial expected behaviour.
