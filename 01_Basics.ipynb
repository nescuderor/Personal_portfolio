## Importing necessary libraries

```python
import os
import numpy as np
import pandas as pd
import scipy
import matplotlib.pyplot as plt
```

## Data simulation

```python
#Creating an array for basic plotting
# np.random.seed(300) #Setting the seed
simulated_data = np.column_stack([np.random.normal(0, 1, 100), np.random.binomial(10, 0.5, 100)]) #Simulating the data with two variables, one discrete and another continuous

#Preview
print(simulated_data[:5,], simulated_data[-5:,], sep="\n\n")
```

```python
#Measurements of central tendency manual calculus
for column in range(0, simulated_data.shape[1]):
	print("Mean: {}, median: {}, variance: {}, skewness: {}, kurtosis: {}".format(	
		np.mean(simulated_data[:,column]),
		np.median(simulated_data[:,column]),
		np.var(simulated_data[:,column]),
		scipy.stats.skew(simulated_data[:,column], axis=0, bias=False),
		scipy.stats.kurtosis(simulated_data[:,column], axis=0, fisher=False, bias=False)))
	
print(f"Correlation coefficient \n{np.corrcoef(simulated_data[:, 0], simulated_data[:, 1])}")
```

According to our random simulated process, and confirmed by the descriptive statistics, the two variables are not linearly correlated.

```python
#Exploratory plot
plt.plot(simulated_data[:, 0], simulated_data[:, 1])
plt.title("Exploratory plot variable 1 vs variable 2")
plt.show()

#Basic plots for understanding their distribution and concentration
fig, axs = plt.subplots(2, 2)

axs[0, 0].set_title('Strip plot for first column')
axs[0, 0].plot(simulated_data[:,0], "ro")
axs[0, 1].set_title('Histogram for first column')
axs[0, 1].hist(simulated_data[:,0], facecolor='r')

axs[1, 0].set_title('Strip plot for second column')
axs[1, 0].plot(simulated_data[:,1], "bo")
axs[1, 1].set_title('Histogram for second column')
axs[1, 1].hist(simulated_data[:,0])

fig.tight_layout()
```
